```python
import pandas as pd

factions = pd.read_html("https://twilight-imperium.fandom.com/wiki/Factions")[0]
players = {"Filip": [], "Josef": [], "Rasmus": [], "Håkan A": [], "Håkan Å": []}

for player in players:
    factions_sample = factions.sample(n=3, random_state=35684)
    players[player] = factions_sample[1]
    factions.drop(factions_sample.index, inplace=True)
   

players = pd.DataFrame(players)
players
```

```python
aliases = pd.DataFrame(
    {
        "Filip": "Kiss3n",
        "Josef": "Magistraten",
        "Rasmus": "eggmcmuffinman",
        "Håkan A": "Fliktor",
        "Håkan Å": "LudwigVan",
    },
    index=["Alias"],
)
pd.concat([aliases, players])
```

```python
expansion_factions = pd.read_html("https://twilight-imperium.fandom.com/wiki/Factions")[1]
players_1 = {"Filip": [], "Josef": [], "Rasmus": [], "Håkan A": [], "Håkan Å": []}

for player in players_1:
    factions_sample = expansion_factions.sample(n=1, random_state=35684)
    players_1[player] = factions_sample[1]
    expansion_factions.drop(factions_sample.index, inplace=True)

players_1 = pd.DataFrame(players_1)
players_1
```

```python
pd.concat([aliases, players, players_1])
```
