```python
df1 = vaccin["Vaccinationer tidsserie"]
```

```python
df1["Veckonummer"] = df1["År"].astype(str) + "v" + df1["Vecka"].astype(str)
px.line(df1, x= "Veckonummer", y="Antal vaccinationer", color="Region", log_y=True)
```

```python
vaccin_kommun_alder = vaccin["Vaccinerade kommun och ålder"]
befolkning_per_kommun = vaccin_kommun_alder.groupby("Län_namn")["Befolkning"].sum()

befolkning_per_kommun = pd.concat([pd.Series({"| Sverige |": befolkning_per_kommun.sum()}), befolkning_per_kommun])
befolkning_per_kommun

df1 = pd.concat([df1, befolkning_per_kommun])
df1.rename(columns={0: "Befolkning"}, inplace=True)
df1

df1["Andel vaccinerade"] = df1["Antal vaccinationer"] / df1["Befolkning"][df1["Region"]]

```
