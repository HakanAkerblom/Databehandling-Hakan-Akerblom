```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

demography = pd.read_html("https://sv.wikipedia.org/wiki/Sveriges_demografi", match="Döda")[0]
demography.rename(columns={"Unnamed: 0": "År"}, inplace=True)
```

```python
demography["Total fertilitet"][0]
```

```python
fig, axes = plt.subplots(1, 3, figsize = (16, 4), dpi = 120)

demography["Medel- folkmängd"] = demography["Medel- folkmängd"].str.replace(" ", "").str.replace(",0", "").str.replace(",5", "").astype(int)
demography["Födda"] = demography["Födda"].str.replace(" ", "").astype(int)
demography["Total fertilitet"] = demography["Total fertilitet"]/100
sns.lineplot(demography, x = "År", y = "Medel- folkmängd", ax=axes[0])
sns.lineplot(demography, x = "År", y = "Födda", ax = axes[1])
sns.lineplot(demography, x = "År", y = "Total fertilitet", ax = axes[2]);

```
